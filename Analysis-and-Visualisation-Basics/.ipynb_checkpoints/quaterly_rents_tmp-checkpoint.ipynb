```python
import pandas as pd
from pathlib import Path
import json
```

```python
src_file = Path.cwd()/'Quarterly median rents by local government area - June quarter 2021.xlsx'
src_file
```

### Perform necessary handling to import the dataset into a pandas dataframe

```python
df_dict = pd.read_excel(src_file, header=[1,2], index_col=[0,1], sheet_name=None)
df_dict
```

```python
excel_sheet_list = list(df_dict.keys())
excel_sheet_list
```

```python
isinstance(df_dict['All Properties'].index, pd.MultiIndex)
# type(df_dict['All Properties'])
```

```python
df1 = df_dict['All Properties']
df1
```

```python
df1_new = df1.stack(0).rename_axis(('Region', 'LGA', 'Quater')).reset_index()
print(df1_new.columns)
print(df1_new.shape)
df1_new.head(15)
```

```python
# Stack columns of data frame to make it single indexed
def stack_df(df):
    df_new = df.stack(0).rename_axis(('Region', 'LGA', 'Quater')).reset_index()
    print(df_new.columns)
    print(df_new.shape)
    print(df_new.head(15))
    return df_new
```

```python
stacked_df_dict = {k:stack_df(v) for k,v in df_dict.items()}
```

```python
df1_new['Count'] = pd.to_numeric(df1_new['Count'], errors='coerce')
df1_new['Count'].dtype
print(df1_new['Count'].max())
print(df1_new['Count'].min())
```

```python
df1_new['Median'] = pd.to_numeric(df1_new['Median'], errors='coerce')
df1_new['Median'].dtype
print(df1_new['Median'].max())
print(df1_new['Median'].min())
```

### Save the extracted dataset in .csv and .json formats

```python
# Write to csv
for sheet in excel_sheet_list:
    print(f"Saving '{sheet}' dataset as")
    csv_file = Path.cwd()/f"{sheet}.csv"
    print(csv_file)
    stacked_df_dict[sheet].to_csv(path_or_buf=csv_file)
```

```python
# Write to json
for sheet in excel_sheet_list:
    print(f"Saving '{sheet}' dataset as")
    json_file = Path.cwd()/f"{sheet}.json"
    print(json_file)
    d = stacked_df_dict[sheet].to_dict()
    with open(json_file, 'w') as f:
        f.write(json.dumps(d, indent=4))
```
